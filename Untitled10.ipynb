```python
fraudData = pd.read_csv("C:\\Users\\julie\\OneDrive\\Bureau\\m1\\s\\SUPPORTS\\data\\fraudData.csv ",  header=0)
fraudData
```

```python
fraudData['signup_time'] = pd.to_datetime(fraudData['signup_time'])
fraudData['purchase_time'] = pd.to_datetime(fraudData['purchase_time'])
fraudData['diff_time'] = (fraudData['purchase_time'] - fraudData['signup_time'])/np.timedelta64(1, 's')
fraudData
```

```python
device_count = fraudData.groupby('device_id')['user_id'].count()
device_user_count = device_count[fraudData['device_id']]
device_user_count = device_user_count.reset_index().drop('device_id', axis = 1)
fraudData['device_user_count'] = device_user_count
```

```python
fraudData
```

```python
ip_count = fraudData.groupby('ip_address')['user_id'].count()
ip_count = ip_count[fraudData['ip_address']].reset_index().drop('ip_address', axis = 1)
ip_count.columns = ['ip_count']
fraudData = pd.concat([fraudData, ip_count], axis = 1)
fraudData
```

```python
y = fraudData['class']
y


```

```python
fraudData = fraudData.drop(['user_id', 'signup_time','purchase_time','class'], axis = 1)
X = fraudData

lb = LabelEncoder()
X['device_id'] = lb.fit_transform(X['device_id'])
X['source'] = lb.fit_transform(X['source'])
X['browser'] = lb.fit_transform(X['browser'])
X['sex'] = lb.fit_transform(X['sex'])
X['country'] = lb.fit_transform(X['country'])
```
